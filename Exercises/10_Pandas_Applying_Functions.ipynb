```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['salary_year_avg']
```

```python
df_salaries = df[pd.notna(df['salary_year_avg'])].copy()
```

```python
df_salaries['salary_year_inflated'] = df_salaries['salary_year_avg'].apply(lambda salary: salary*1.03)
df_salaries[['salary_year_avg', 'salary_year_inflated']]
```

```python
import ast

ast.literal_eval(df['job_skills'][1])
```

```python
def clean_list(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)
```

```python
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
def projected_salary(row): 
    if "Senior" in row['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else: 
        return 1.03 * row['salary_year_avg']

df_salaries['salary_year_inflated'] = df_salaries.apply(projected_salary, axis=1)
df_salaries[['job_title_short', 'salary_year_avg', 'salary_year_inflated']].sample(10)
```

```python
df_salaries['salary_year_inflated'] = df_salaries.apply(lambda row: 1.05*row['salary_year_avg'] if "Senior" in row['job_title_short'] else 1.03*row['salary_year_avg'], axis=1)
df_salaries[['job_title_short', 'salary_year_avg', 'salary_year_inflated']].sample(10, random_state=42)
```
